# Data cleaning

```python
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
```

```python
stopword_set = list(stopwords.words('english'))
stopword_set[:10]
```

```python
text = train_docs[0]
text
```

```python
# remove punctuation and numbers
import re
clean_text = re.sub(r"\n|(\\(.*?){)|}|[!$%^&*#()_+|~\-={}\[\]:\";'<>?,.\/\\]|[0-9]|[@]", ' ', text)
clean_text
```

```python
# remove extra space
clean_text = re.sub('\s+', ' ', clean_text)
clean_text
```

```python
clean_text_list = [word.lower() for word in clean_text.split()] # to list 
clean_text_list
```

```python
#remove stop words 
clean_text_list = [word for word in clean_text_list if word not in stopword_set] 

len(clean_text_list)
```
